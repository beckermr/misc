```python
import galsim
import numpy as np
import matplotlib.pyplot as plt
```

```python
obj = galsim.Gaussian(fwhm=0.9)
```

```python
cen = galsim.PositionD(45.2, 66.7)
```

```python
dim = 50
img_even = obj.drawImage(nx=dim, ny=dim, center=cen, scale=0.2)
img_odd = obj.drawImage(nx=dim+1, ny=dim+1, center=cen, scale=0.2)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

vmin = np.min(img_even.array)
vmax = np.max(img_even.array)

axs[0].imshow(img_even.array, vmin=vmin, vmax=vmax)
axs[1].imshow(img_odd.array, vmin=vmin, vmax=vmax)
axs[2].imshow(img_odd.array[:-1, :-1] - img_even.array, vmin=vmin, vmax=vmax)
```
