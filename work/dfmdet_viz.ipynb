```python
import proplot as pplt

from deep_field_metadetect.utils import canned_viz_for_obs, make_simple_sim
```

```python
res = make_simple_sim(seed=10, n_objs=20, dim=151)
```

```python
canned_viz_for_obs(res[1])
```

```python
import numpy as np

fig, axs = pplt.subplots(nrows=1, ncols=2, figsize=(6, 3))

axs[0].imshow(np.arcsinh(res[0].image * np.sqrt(res[0].weight)), cmap="rocket")
axs[0].format(grid=False, xlabel="x [pixels]", ylabel="y [pixels]", title="wide-field image")

axs[1].imshow(np.arcsinh(res[1].image * np.sqrt(res[1].weight)), cmap="rocket")
axs[1].format(grid=False, title="deep-field image")
```
